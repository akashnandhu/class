```python
import pyautogui
import time

message = "haiii"

print("You have 5 seconds to click on the WhatsApp chat...")
time.sleep(5)

for i in range(700):
    pyautogui.typewrite(message)
    pyautogui.press("enter")

print("Done! Sent 400 messages.")

```
